#Image to Pencil Sketch(LGM)

### Import Dependencies

```python
#READ AN IMAGE
import cv2
import numpy as np
from matplotlib import pyplot as plt

import warnings
warnings.filterwarnings('ignore')
img = cv2.imread("/content/Elon_Musk_Img.jpg")
#SEE THE TYPE OF IMAGE i.e AN NUMPY ARRAY
print(type(img))
#TO SEE THE IMAGE
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
color = ('b','g','r')
for i ,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.show()
```

```python
#convert bgr to rgb
img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
plt.axis(False)
plt.show()
```

### convert an image from BGR to RGB

```python
color = ('b','g','r')
for i ,col in enumerate(color):
    histr = cv2.calcHist([img_RGB],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.show()
```

### Convert RGB to GrayScale

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
plt.axis(False)
plt.show()
```

### Inverting the image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.imshow(img_invert)
```

### Blur the image

```python
img_blur = cv2.GaussianBlur(img_invert,(111,111),0)
plt.imshow(img_blur)
```

### Inverting Blur Image

```python
invert_blur_img = cv2.bitwise_not(img_blur)
plt.imshow(invert_blur_img)
```

### Convert it to Sketch Image

```python
img_sketch = cv2.divide(img_gray,invert_blur_img,scale=256.0)
plt.imshow(img_sketch)
```

### Convert sketch to BGR

```python
plt.figure(figsize = (7,7))
img2 = cv2.cvtColor(img_sketch,cv2.COLOR_RGB2BGR)
plt.imshow(img2)
plt.axis(False)
plt.show()
```
